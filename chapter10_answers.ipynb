# Q1: Number Guessing Game

**Pseudocode:**

1. Import the "random" module for generating random numbers.

2. Set the lower limit and upper limit for the random number:

    lower_limit = 1
    upper_limit = 100

3. Generate a random number within the specified range and store it as the secret_number using `random.randint(lower_limit, upper_limit).`

4. Set the maximum number of attempts allowed:

    max_attempts = 10
    Initialize attempts to 0.

5. Display a welcome message to the user, including the number of attempts and the range of numbers.

6. Start a loop that continues until attempts is less than max_attempts:

    a. Inside the loop, try to do the following:

        - Get the user's guess by taking input as an integer and store it in the variable guess.
        - Increment attempts by 1.
        - Display the user's guess and the number of attempts left.

        - Check if the user's guess is equal to the secret_number:
            -- If they guessed correctly, display a congratulations message with the correct number and the number of attempts taken.
            -- Exit the loop using "break."

        - If the guess is not correct, check if it's lower or higher than the secret_number and provide appropriate hints.

        - If the user enters invalid input (not a valid number), display an error message and prompt them to enter a valid number.

7. If the loop exits without the user guessing the correct number (i.e., attempts equals max_attempts), display a message indicating that the maximum number of attempts has been reached and reveal the secret_number.

8. End the program.

```python
# Q1
import random

# Set the range for the random number
lower_limit = 1
upper_limit = 100

# Generate a random number within the specified range
secret_number = random.randint(lower_limit, upper_limit)

# Set the number of attempts allowed
max_attempts = 10
attempts = 0

print(f"Welcome to the Number Guessing Game! You have {max_attempts} attempts to guess the number between {lower_limit} and {upper_limit}.")

while attempts < max_attempts:
    try:
        # Get the user's guess
        guess = int(input("Enter your guess: "))
        
        # Increment the number of attempts
        attempts += 1

        print(f"You guessed: {guess} || Attempts left: {max_attempts-attempts}")
        
        # Check if the guess is correct
        if guess == secret_number:
            print(f"Congratulations! You guessed the correct number {secret_number} in {attempts} attempts.")
            break
        elif guess < secret_number:
            print("Try a higher number.")
        else:
            print("Try a lower number.")
        
    except ValueError:
        print("Invalid input. Please enter a valid number.")

if attempts == max_attempts:
    print(f"Sorry, you've reached the maximum number of attempts. The secret number was {secret_number}.")
```

# Q2: Word Jumble Game

Pseudocode:
![Alt text](image-1.png)

```python
# Q2
import random

# List of words for the game
word_list = ["python", "java", "programming", "computer", "algorithm", "coding"]

# Select a random word from the list
random_word = random.choice(word_list)

# Create a jumbled version of the word
jumbled_word = list(random_word)
random.shuffle(jumbled_word)
jumbled_word = ''.join(jumbled_word)

# Number of chances for the player
attempts = 3

print("Welcome to the Word Jumble Game!")
print("Unscramble the letters to guess the word.")
print(f"You have {attempts} attempts to guess the word.")

while attempts > 0:
    print("\nJumbled word:", jumbled_word)
    guess = input("Your guess: ").lower()

    print(f"You guessed: {guess}")
    if guess == random_word:
        print("Congratulations! You guessed the word correctly.")
        break
    else:
        print("Incorrect guess.")
        attempts -= 1
        if attempts > 0:
            print(f"You have {attempts} attempts left.")
        else:
            print("You've run out of attempts. The word was:", random_word)
```

```python
import random

player1_name = input("Enter Player 1's name: ")
player2_name = input("Enter Player 2's name: ")

player1_score = random.randint(1, 6) + random.randint(1, 6)
player2_score = random.randint(1, 6) + random.randint(1, 6)

print(f"{player1_name} rolled a {player1_score}")
print(f"{player2_name} rolled a {player2_score}")

if player1_score > player2_score:
    print(f"{player1_name} wins!")
elif player2_score > player1_score:
    print(f"{player2_name} wins!")
else:
    print("It's a tie!")
```

# Q4: Shop Inventory System

```python
# Initialize lists to store item names, prices, and quantities
item_names = ["Apple", "Banana", "Orange"]
item_prices = [0.5, 0.25, 0.75]
item_quantities = [10, 20, 15]

while True:
    # Display menu
    print("\nShop Inventory System")
    print("1. Add an item")
    print("2. List all items")
    print("3. Update stock")
    print("4. Remove an item")
    print("5. Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        # Add an item
        name = input("Enter item name: ")
        price = float(input("Enter item price: "))
        quantity = int(input("Enter item quantity: "))
        
        item_names.append(name)
        item_prices.append(price)
        item_quantities.append(quantity)
        print(f"{name} added to the inventory at price ${price} and quantity {quantity}.")

    elif choice == "2":
        # List all items
        print("\nInventory List:")
        for i in range(len(item_names)):
            print(f"Item: {item_names[i]}, Price: ${item_prices[i]}, Quantity: {item_quantities[i]}")

    elif choice == "3":
        # Update stock
        name = input("Enter item name to update: ")
        if name in item_names:
            index = item_names.index(name)
            new_quantity = int(input("Enter new quantity: "))
            item_quantities[index] = new_quantity
            print(f"{name}'s quantity updated to {new_quantity}.")
        else:
            print(f"{name} not found in the inventory.")

    elif choice == "4":
        # Remove an item
        name = input("Enter item name to remove: ")
        if name in item_names:
            index = item_names.index(name)
            item_names.pop(index)
            item_prices.pop(index)
            item_quantities.pop(index)
            print(f"{name} removed from the inventory.")
        else:
            print(f"{name} not found in the inventory.")

    elif choice == "5":
        # Quit the program
        print("Exiting the inventory system. Goodbye!")
        break

    else:
        print("Invalid choice. Please select a valid option.")
```

# Q5: ATM

```python
# Initialize the account balance
balance = 1000

# Display the welcome message and menu
print("Welcome to the ATM!")
print(f"Current Balance: ${balance}\n")

while True:
    # Display the menu options
    print(40*"-3")
    print("Menu:")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")
    
    # Get user choice
    choice = input("\nEnter your choice: ")
    
    if choice == '1':
        # Deposit
        deposit_amount = float(input("Enter the amount to deposit: "))
        balance += deposit_amount
        print(f"Deposit successful. New Balance: ${balance}\n")
    elif choice == '2':
        # Withdraw
        withdraw_amount = float(input("Enter the amount to withdraw: "))
        if withdraw_amount <= balance:
            balance -= withdraw_amount
            print(f"Withdrawal successful. New Balance: ${balance}\n")
        else:
            print(f"Error: Insufficient funds. Current Balance: ${balance}\n")
    elif choice == '3':
        # Check Balance
        print(f"Current Balance: ${balance}\n")
    elif choice == '4':
        # Exit
        print("Thank you for using the ATM!")
        break
    else:
        print("Invalid choice. Please select a valid option.\n")
```
